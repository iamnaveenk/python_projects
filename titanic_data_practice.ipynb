```python
import pandas as pd
```

```python
df=pd.read_csv("https://bit.ly/3kGDuKx", na_values="?")
```

```python
df.head()
```

```python
df.groupby(['pclass'])['survived'].aggregate(['mean','count','sum','min','max'])
```

```python
df.groupby(['pclass'])['survived'].agg(['mean','count','sum','min','max'])
```

```python
df.groupby(['survived'])['pclass'].agg(['mean','count','sum','min','max']).reset_index()
```

```python
df.groupby(['pclass'])['survived','fare'].sum()
```

```python
df.groupby(['pclass','sex'])['survived'].sum()
```

```python
df.groupby(['pclass'],as_index=False).sum()
```

```python
df['age_bins']=pd.cut(df['age'],bins=3, labels=('young','middle_age','old'))
```

```python
df.head()
```

```python
df.groupby(['age_bins'])['survived'].sum()
```
